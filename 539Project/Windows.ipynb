```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sc
from scipy.stats import gamma
```

```python
df = pd.read_csv('flukaRunsTyped.csv',dtype='float64')
data = df.to_numpy()
```

```python
#create dictionary for easy access
energies = [10,30,70,100,300,700,1000,3000,7000,10000]
particle_types = ['PION','PION-','KAONLONG','PROTON','APROTON','NEUTRON']
dct = {}
for particle in particle_types:
    dct[particle] = {}
    for energy in energies:
        dct[particle][energy] = []

for i in range(len(df)):
    for particle in particle_types:
        if df[particle][i]:
            dct[particle][df.energy[i]].append(data[i,0:500])
        
```

```python
#Params
binLim = 150
```

```python
energy = 10
xvals = np.arange(500)[0:binLim]
ax = plt.gca()
for particle in particle_types:
    yvals = np.mean(dct[particle][energy],axis=0)[0:binLim]
    yerr = np.std(dct[particle][energy],axis=0)[0:binLim]
    color=next(ax._get_lines.prop_cycler)['color']
    plt.plot(10 * xvals, yvals,label=particle,color=color,linestyle='solid')
    plt.plot(10 * xvals, yvals-yerr,color=color,linestyle='dashed')
    plt.plot(10 * xvals, yvals+yerr,color=color,linestyle='dashed')
    plt.fill_between(10 * xvals, yvals-yerr, yvals+yerr,color=color,alpha=0.1)
plt.title('Average longitudinal profile for various particles (' + str(energy) + ' GeV)')
plt.xlabel('Distance from beginning of shower (cm)')
plt.ylabel('Weighted track length (cm)')
plt.legend()
plt.yscale('linear')
plt.show()
```
